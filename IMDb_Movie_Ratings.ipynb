# **Exploratory Data Analysis (EDA)**

```python
#Finding null values
df.isnull().sum()
df_null = df.isnull().sum().any()
print("Null Values:", df_null)
```

```python
sns.heatmap(df.isnull())
```

```python
df = df.dropna(axis=0)
df_null = df.isnull().sum().any()
print("Null Values:", df_null)
```

```python
sns.heatmap(df.isnull())
```

```python
#Checking to see if there are any duplicate rows
df.duplicated()
duplicate_data = df.duplicated().any()
print("Duplicated vaues:", duplicate_data)
```

```python
#Statistics
df["Rating"].describe(include='all')
```

```python
df.columns
```

```python
#Distribution of Movie Ratings
plt.figure(figsize = (8,6))
sns.histplot(df['Rating'], bins = 20, kde = True)
plt.title("Distribution of Movie Ratings")
plt.xlabel("IMDb Ratings")
plt.ylabel("Frequency")
```

```python
#Average Rating for each Director
df.groupby("Director")["Rating"].mean().sort_values(ascending = False)
```

```python
#Total Runtime per Movies
top10_len = df.nlargest(10, "Runtime (Minutes)")[["Title", "Runtime (Minutes)"]]\
.set_index("Title")
```

```python
sns.barplot(x = "Runtime (Minutes)", y = top10_len.index, data = top10_len)
plt.title("Total Runtime per Movies")
plt.show()
```

```python
#Number of Movies per Year
top10_len = df.nlargest(10, "Runtime (Minutes)")[["Title", "Runtime (Minutes)"]]\
.set_index("Title")
```

```python
sns.countplot(x = "Year", data = df)
plt.title("Number of Movies per Year")
plt.show()
```

```python
#Top 10 Highest Rated Movie Titles and its Directors
top10_mov = df.nlargest(10, "Rating")[["Title", "Rating", "Director"]]\
.set_index("Title")
top10_mov
```

```python
sns.barplot(x = "Rating", y = top10_mov.index, data = top10_mov, hue = "Director", dodge = False)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2)
plt.title("Top 10 Highest Rated Movie Titles and its Directors")
plt.show()
```

```python
#Top 10 Genres
top10_gen = df.nlargest(10, "Rating")[["Genre", "Rating"]]\
.set_index("Genre")
top10_gen
```

```python
sns.barplot(x = "Rating", y = top10_gen.index, data = top10_gen, hue = "Genre")
plt.title("Top 10 Highest Rated Genres")
plt.show()
```

```python
#Dashboard
fig, axes = plt.subplots(2, 2, figsize = (20, 16))
sns.histplot(df['Rating'], bins = 20, kde = True, ax = axes[0, 0])
axes[0, 0].set_title("Distribution of Movie Ratings")
sns.barplot(x = "Runtime (Minutes)", y = top10_len.index, data = top10_len, ax = axes[0, 1])
axes[0, 1].set_title("Total Runtime per Movies")
sns.barplot(x = "Rating", y = top10_gen.index, data = top10_gen, hue = "Genre", ax = axes[1, 0])
axes[1, 0].set_title("Top 10 Highest Rated Genres")
sns.barplot(x = "Rating", y = top10_mov.index, data = top10_mov, hue = "Director", dodge = False, ax = axes[1, 1])
axes[1, 1].legend(bbox_to_anchor = (1.05, 1), loc = 2)
axes[1, 1].set_title("Top 10 Highest Rated Movie Titles and its Directors")
plt.tight_layout()
plt.show()
```
